```python
import csv
import pandas as pd
import numpy as np
```

```python
def check_packet_intrusion(row):
  if row[2] == row[3]:
    return True
  elif row[2] == '255.255.255.255':
    return True
  elif row[3] == '255.255.255.255':
    return True
  elif row[5] < 40:
    return True
  elif row[15] == 'ICMP':
    return True
  else:
    return False
```

```python
def checkFile(filename, outFileName):
  df = pd.read_csv(filename)
  df_col = df.columns
  intrusion_count = 0
  count = 0
  txtfile = open(outFileName,"w")
  for index, row in df.iterrows():
    if row.isnull().values.any():
      continue
    count += 1
    if check_packet_intrusion(row) == True:
      print("Analyzed Packet is Intrusion Packet")
      txtfile.write("Analyzed Packet is Intrusion Packet\n")
      intrusion_count += 1
    else:
      print("Analyzed Packet is Not-Intrusion Packet")
      txtfile.write("Analyzed Packet is Not-Intrusion Packet\n")
    for i in range(len(row)):
      print("{0} : {1} ".format(df_col[i],row[i]), end =" ")
      txtfile.write("{0} : {1} ".format(df_col[i],row[i]))
    txtfile.write('\n')
    print("\n")
  print("total number of packets checked is = {0}".format(count))
  txtfile.write("total number of packets checked is = {0} \n".format(count))
  print("total number of intrusion packets detected is = {0}".format(intrusion_count))
  txtfile.write("total number of intrusion packets detected is = {0}".format(intrusion_count))
  txtfile.close()
```

```python
checkFile("Sample-Testcase-1.csv", "Sample-Testcase-1_output.txt")
```
